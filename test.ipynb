```python
import numpy as np
import pandas as pd

# Carregar os datasets
data_1 = cosmic.ler_arquivo_dat("utils/C1_CP_FGM_FULL/data.dat")
data_2 = cosmic.ler_arquivo_dat("utils/C2_CP_FGM_FULL/C2_CP_FGM_FULL__20010213_104200_20010213_105000_V140306.cef")
data_3 = cosmic.ler_arquivo_dat("utils/C3_CP_FGM_FULL/C3_CP_FGM_FULL__20010213_104200_20010213_105000_V140305.cef")
data_4 = cosmic.ler_arquivo_dat("utils/C4_CP_FGM_FULL/C4_CP_FGM_FULL__20010213_104200_20010213_105000_V140305.cef")

# Extrair valores relevantes dos datasets
```

```python
data_2.columns = data_1.columns
data_3.columns = data_1.columns
data_4.columns = data_1.columns
```

```python
for col in data_1.columns:
    try:
        data_1[col] = data_1[col].astype(float)
        data_2[col] = data_2[col].astype(float)
        data_3[col] = data_3[col].astype(float)
        data_4[col] = data_4[col].astype(float)
    except:
        pass
```

```python
data_1 = data_1[:len(data_2)]
data_1
```

```python

curl = cosmic.curlometer(data_1, data_2, data_3, data_4)
curl.plot()
```

```python

# Identify the columns
Bx_column = "B_vec_xyz_gse__C1_CP_FGM_FULL1"
By_column = "B_vec_xyz_gse__C1_CP_FGM_FULL2"
Bz_column = "B_vec_xyz_gse__C1_CP_FGM_FULL3"



# Calculate mod_B
yy = cosmic.calculate_mod_B(data_1, Bx_column, By_column, Bz_column)

# Plot mod_B
cosmic.plot_mod_B(yy)

# Save mod_B to a file
#yy.to_csv('mod_B.dat', sep='\t', header=False, index=False)

PVI = cosmic.calculate_PVI(yy)

# Plot the PVI
cosmic.plot_data(PVI)

```

```python

current_density = np.loadtxt('utils/current_sheet/current_density.dat', converters={0: cosmic.convert_to_float})
```

```python

volat = cosmic.calculate_magnetic_volatility(pd.DataFrame(yy, columns=["mag"]), "mag", 100, 100)
volat.plot()
```

```python

sigma = 50.0

smoothed_curl = cosmic.apply_gaussian_kernel(current_density, sigma)
smoothed_pvi = cosmic.apply_gaussian_kernel(PVI, sigma)
smoothed_volat = cosmic.apply_gaussian_kernel(volat, sigma)

```

```python
import matplotlib.pyplot as plt
plt.plot(PVI, color="black", label="PVI")
plt.plot(smoothed_pvi, color="red", label="PVI suavizada")
plt.xlabel('Tempo')
plt.ylabel('PVI')
plt.legend()
# Set the title of the plot
plt.title('Filtro Gaussiano Sigma = 50.0')


plt.show()
```
